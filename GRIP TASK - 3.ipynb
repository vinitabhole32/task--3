```python
plt.figure(figsize=(12,12))
plot=data.groupby(['State'])['Profit'].sum().nlargest(50)
plot.plot.bar()
```

**Graph shows the Highest & Lowest Profit according to State**

```python
plt.style.use('seaborn')
data.plot(kind='scatter',figsize=(20,8),x='Sales',y='Profit',c='Discount',s=20,fontsize=12,colormap='plasma')
plt.ylabel('Profits')
plt.xlabel('Sales')
plt.show()
```

```python
sns.lineplot(x='Discount',y='Profit',label='Profit',data=data)
plt.legend()
plt.show()
```

**Here the relation between profit and discount is of negative type. As discount increases profit decreases**

```python
a=data.groupby('Region')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
a[:].plot.bar(color=['blue','green'],figsize=(12,6))
plt.title('Profit/Loss & Sales across the region')
plt.xlabel('Region')
plt.ylabel('Profit/Loss & Sales')
plt.show()
```

**Here we observed that when we increase discount the sales increases but profit decreases. And West region has higest sales & profit**

```python
a=data.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
a[:].plot.bar(color=['blue','green'],figsize=(20,10))
plt.title('Profit/Loss & Sales across the states')
plt.xlabel('States')
plt.ylabel('Profit/Loss & Sales')
plt.show()
```
